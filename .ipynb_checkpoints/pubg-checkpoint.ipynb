# Categorical features
## Look for numeric values that behave as if categorical -- scratched for first round of predictions; just treat as numerical.. bucket a few

```python
# look for numerical values that act as cat
actual_cat=[]
actual_int=[]
merged_ints=mergeddf[mergeddf.dtypes[(mergeddf.dtypes!='object')&(mergeddf.dtypes!='bool')].index]

for col in list(merged_ints):
    if merged_ints[col].nunique()*1.0/merged_ints.shape[0] <= 0.2: # if the number of unique values is less than 20%
        actual_cat.append(col)
    else:
        actual_int.append(col)

print 'actually_categorical'
print actual_cat
print 'actually numeric'
print actual_int
```

```python
sns.set(font_scale=1.25)

print mergeddf[actual_cat].describe()
mergeddf[actual_cat].hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(15,20)
plt.suptitle("Distribution of Numeric Features")
plt.show()
```

# Bucket Numeric Features Based on Percentile
Create describe dataframes for each numeric column. Loop through each numeric column and assign percentile rank to temporary list, once all numbers are assigned rank append to a master list. create dataframe from master list then merge to the main dataframe

```python
numeric_buckets=[]
for i in actual_cat:
    numeric_buckets.append(mergeddf[i].describe())
```

```python
percentile_rank=[]
for i in actual_cat:
    temp_list=[]
    column_to_list=mergeddf[i].tolist()
    for c in column_to_list:
        if c <= numeric_buckets[actual_cat.index(i)]['25%']:
            temp_list.append(1)
        elif c <= numeric_buckets[actual_cat.index(i)]['50%']:
            temp_list.append(2)
        elif c <= numeric_buckets[actual_cat.index(i)]['75%']:
            temp_list.append(3)
        else:
            temp_list.append(4)
    percentile_rank.append(temp_list)
    
```

```python
num_rank_df=pd.DataFrame()
for i in actual_cat:
    num_rank_df['rank_'+i]=percentile_rank[actual_cat.index(i)]

num_rank_df.head()
```

```python
#Validation function
n_folds = 5

#function to evaluate each model
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42)#.get_n_splits(X)
    rmse= np.sqrt(-cross_val_score(model, X.sample(random_state=1,n=50000), Y.sample(random_state=1,n=50000), scoring="neg_mean_squared_error", cv = kf, verbose=True))
    return(rmse)
```
